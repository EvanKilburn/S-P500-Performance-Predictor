* GBP, JPY, CAD, CNY - change in currency (decimal of percent change from previous day)

* CTB3M,CTB6M,CTB1Y - change in Canadian T-Bill of varying durrations (decimal of percent change from previous day)

## Training
### Training Description
We will use supervised learning which is a type of machine learning that trains itself using known inputs and outputs to generalize future outputs. A distinct feature of supervised learning when compared to unsupervised learning is that it knows the desired output for a given input (the data set is labeled). We will be using linear regression which helps predict a continuous quantity where as classification which predicts a distinct class label. For our use case we are trying to predict a continuous quantity which is the value of the S&P 500 dollar value of the next day's closing price. Linear regression can achieve this as it is not a distinct class. We will be using Amazon's Linear Learner models algorithm which is a supervised machine learning algorithm that can handle both classification and regression. We will be using the regression capabilities to predict the future close the the S&P 500.

### Training the Model

```python
# Shuffle and Split the data into 80% training and 20% testing

train_data, test_data = train_test_split(market_data, test_size=0.2)

# Convert dataframes to back to CSV files and save them locally to the notebook
# Make sure to exclude the headers so that Linear Learner will accept the file

train_data.to_csv('train.csv', header=False, index=False)
test_data.to_csv('test.csv', header=False, index=False)
```

```python
# Copy the file to a S3 bucket using sess.upload_data
# We will need to pass the file to be uploaded, a prefix (top level folder to store the data) and the subfolders
# Here we will also define the output folder for the results

prefix = 'market'
training_data_path = sess.upload_data( path='train.csv', key_prefix=prefix + '/input/train') 
testing_data_path = sess.upload_data( path='test.csv', key_prefix=prefix + '/input/test')
output_data_path = 's3://{}/{}/output'.format(bucket, prefix)
print(training_data_path)
print(testing_data_path)
print(output_data_path)
```

```python
# specify containers

from sagemaker.image_uris import retrieve


# List of frameworks that can be retrieved https://sagemaker.readthedocs.io/en/stable/frameworks/index.html
# list of Built-in Algorithms that can be retrieved https://sagemaker.readthedocs.io/en/stable/algorithms/index.html

container = retrieve('linear-learner', region)
```

```python
# Amazon Estimator Documentation https://sagemaker.readthedocs.io/en/stable/algorithms/sagemaker.amazon.amazon_estimator.html
# Amazon's Estimator base class

from sagemaker.estimator import Estimator 
ll_estimator = Estimator(container, 
                         role=role, 
                         instance_count=1, 
                         instance_type='ml.c4.xlarge', 
                         output_path=output_data_path)
```

```python
# Set the linear learner hyperparameters
# hyperparameters can be found here https://docs.aws.amazon.com/sagemaker/latest/dg/ll_hyperparameters.html
# Here we will specify the Hyperameters values
# specifying predictor_type to be of type 'regressor' as we are doing linear regression
# specifying loss to be 'squared_loss' to minimize the squared loss
# specifying normalize_label to 'true' which shifts the label to have a mean of zero and scales it to have unit standard deviation
# specifying unbias_label to 'true' which unbiases labels before training so that the mean is 0

ll_estimator.set_hyperparameters(predictor_type='regressor', 
                                 loss='squared_loss', 
                                 normalize_label='true', 
                                 unbias_label='true')
```

```python
# Creating a channel from the S3 data source and file system, which is where we are storing our training and testing data
# TrainingInput documentation https://sagemaker.readthedocs.io/en/stable/api/utility/inputs.html

from sagemaker import TrainingInput 
training_data_channel = TrainingInput( s3_data=training_data_path, content_type='text/csv') 
testing_data_channel = TrainingInput( s3_data=testing_data_path,   content_type='text/csv')
```

```python
# Hyperparameters = https://docs.aws.amazon.com/sagemaker/latest/dg/ll_hyperparameters.html
# Objectives = https://docs.aws.amazon.com/sagemaker/latest/dg/linear-learner-tuning.html

# set test hyperparameter objective to minimize MSE (Mean Squared Error)

from sagemaker.tuner import IntegerParameter, CategoricalParameter, ContinuousParameter, HyperparameterTuner

hypertuner = HyperparameterTuner(ll_estimator, 
                                 objective_metric_name = 'test:mse',
                                 hyperparameter_ranges = {
                                 'learning_rate' : ContinuousParameter(0.01, 0.1),
                                 'mini_batch_size': CategoricalParameter([32,64,128,256,512])},
                                 objective_type = 'Minimize',
                                 metric_definitions = [{'Name': 'test:mse', 'Regex': 'loss (\S+)'}],
                                 max_jobs=10, 
                                 max_parallel_jobs=1
                                )
```

```python
# Fit this Estimator on our data stored in S3.
# Now we are going to Fit the model

hypertuner.fit( {'train': training_data_channel, 'test': testing_data_channel})
```

```python
print("Best hyerparameter tuning job:{}".format(hypertuner.best_training_job()))
print("Best parameters found={}".format(hypertuner.best_estimator().hyperparameters()))
```

```python
# Set the linear learner hyperparameters
# We will use the tuned values recommended above

ll_estimator.set_hyperparameters(feature_dim='auto',
                                 predictor_type='regressor',
                                 mini_batch_size=64,
                                 learning_rate=0.01,
                                 normalize_label='true',
                                 unbias_label='true',
                                 loss='squared_loss',
                                 _tuning_objective_metric='test:mse')
```

```python
# Now we are going to Fit the model (train it)
# Documentaiton for FIT https://sagemaker.readthedocs.io/en/stable/algorithms/sagemaker.amazon.amazon_estimator.html#sagemaker.amazon.amazon_estimator.AmazonAlgorithmEstimatorBase.fit

ll_estimator.fit( {'train': training_data_channel, 'test': testing_data_channel})
```

### Hyperparameter Tuning
For hyperparameter tuning above, I started off by specifying that we will be using regression as I set the predictor_type to regressor. I specified normalize_label to true and unbias_label to true, to reduce any biases in the data. This will shift the label to have a mean of zero and scale it to have unit standard deviation. As well, above I specified to use squared loss and mean square error for our objective to minimize throughout training and testing. I have chosen to use MSE (Mean Squared Error) instead of absolute loss/MAE (Mean Absolute Error), as it will give higher penalties to values that are off by larger amounts. This will be beneficial to optimize the accuracy of our model. I then ran the hyperparameter tuning using hypertuner.fit and got the following as the recommended parameters:

Best parameters found={'_tuning_objective_metric': 'test:mse', 'learning_rate': '0.01', 'loss': 'squared_loss', 'mini_batch_size': '64', 'normalize_label': 'true', 'predictor_type': 'regressor', 'unbias_label': 'true'}

I took this information and set the ll_estimator hyperparameters accordingly.

Metrics Description:
* MSE (Mean Squared Error) - how close a regression line is to the data, it measures the distance between the data point and the regression line, square it to remove any negative symbols and gives more weight to larger differences, and does this for all data points and then divides by the amount of data points to get the average
* RMSE (Root Mean Square Error) - square root of MSE, it removes the squared units to give a better comparison to the target
* Absolute Loss - the average absolute error between a datapoint and the regression line, and different than RMSE as there is no square root to give higher weight to more incorrect values
* MAE (Mean Absolute Error) - same as absolute loss
* R2 (R Squared) - how well the regression model explains the observed data and its variability (multiply by 100% to get percent value), higher is better
    
Here we can see the testing had a slightly higher MSE (Mean Squared Error) of 595 when compares to the training's MSE of 467. The test error works out to an approximat RMSE of 24.4 (square root of MSE). The RMSE helps determine how far from the regression line our data points are, and with a value of 24.4 this is not bad as our set has a much larger standard deviation of 427.918002. We have an absolute loss/MAE of 18.35 which means on average our prediction was off by approximately 18.35. This is good as it is a relatively low MAE. In contrast to our RMSE of 24.4 we can see how it is higher as values further away were given larger weight, however it is not significantly larger which would indicate that there are outliers and occasional unexplained instances with our model. Last our R2 when converted to a percent is 99.6%. This is an extremely great R2 value as higher is better and it shows that the variance in our data is explained by our input.
    
When evaluating this model there is nothing of significant concern. A future improvement to reduce the RMSE and MAE would be to collect more data to train it on. The best use for this model would as an indicator to compare the estimated close to where the market currently is to determine whether the market is over or under valued based on the historical data it was trained on. This is because with a MAE of 18.35, you could pottentially lose $18.35/share (based on the S&P 500 price, would need to be converted for shares like VOO or VFV.TO). This being used as an indicator would be beneficial as it would help one buy based on long term trends and focus on long term investing. To improve the model's accuracy more data would be beneficial, especially more recent data would allow the model to be more current.

```python
# create unique name as date time appended at the end
sp500_market_epc_name = "linear-learner-serverless-epc" + strftime("%Y-%m-%d-%H-%M-%S", gmtime())

# create an endpoint configuration that is connected to the model
endpoint_config_response = client.create_endpoint_config(
    EndpointConfigName=sp500_market_epc_name,
    ProductionVariants=[
        {
            "VariantName": "llVariant",
            "ModelName": deployed_linear_model_name,
            "ServerlessConfig": {
                "MemorySizeInMB": 4096,
                "MaxConcurrency": 1,
            },
        },
    ],
)

# print unique url (id) for the resource
print("Endpoint Configuration Arn: " + endpoint_config_response["EndpointConfigArn"])
```

```python
# create unique name as date time appended at the end
sp500_market_ep_name = "linear-learner-serverless-ep" + strftime("%Y-%m-%d-%H-%M-%S", gmtime())

# create an endpoint for the model with the above config so we can call the model using this endpoint
create_endpoint_response = client.create_endpoint(
    EndpointName=sp500_market_ep_name,
    EndpointConfigName=sp500_market_epc_name,
)

# print unique url (id) for the resource
print("Endpoint Arn: " + create_endpoint_response["EndpointArn"])
```
